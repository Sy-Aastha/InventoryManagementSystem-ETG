```python
# step 0: adding product to inventory product list
import json
fdr = open("ProductRecord.json",'r')
cT = fdr.read()
fdr.close()
myproducts_record=json.loads(cT)
# print(myproducts_record)

prod_id =input("Enter product id:")
name = str(input("Enter name:"))
sp = int(input("Enter price:"))
qn = int(input("Enter quantity:"))
units=str(input("Enter units[kg/L/pcs]:"))
exp_date=str(input("Enter exp-date:"))
sale=False       #initial sales in false
saleqnt=0
saleTime=[]
transectionID=[]

myproducts_record[prod_id] = {"name": name,"SP":sp,"quant":qn,"units":units,"exp-date":exp_date,
                              "salerec":{"sale":sale,"Salequant":saleqnt,"saleTime":saleTime,"transectionID":transectionID}}
kl= json.dumps(myproducts_record,indent=1)

fd = open("ProductRecord.json",'w')
fd.write(kl)
fd.close()

print("+++++ PRODUCT ADDEDED TO LIST SUCCESSFULLY ++++")

```

```python
# step 1:customer login in/signup and catalogue display
#customer record updates
fdcr = open("Cust_Record.json",'r')
r = fdcr.read()
fdcr.close()
cust_rec = json.loads(r)




nama = input("Enter the first name: ")
pn = input("Enter phno ")
cust_ID=nama+pn[:4]

for x in cust_rec.keys():
    if cust_rec[x]["name"]==nama:
        if cust_rec[x]["Ph.no"]==pn:
            print("Welcome Back!!  Mr/Mrs "+cust_rec[x]["name"] +"   (existing customer)")
            print("Cutomer Id: "+x)
            currentCustID=x
            print("Your previous purchase items[LIST] : "+ str(cust_rec[x]["purchase_rec"]))
            print("")
#             cust_rec[x]["topr"].append("111111111")
            break
    else:
        cust_rec[cust_ID]={"name":nama,"Ph.no":pn,"purchase_rec":[],"topr":[],"transectio_id":[],"billAmount":[]}
        cr = json.dumps(cust_rec,indent=1)
        fdc = open("Cust_Record.json",'w')
        fdc.write(cr)
        fdc.close()
        print("Welcome !!  Mr/Mrs  "+nama+"   (newcustomers)")
        currentCustID=cust_ID
        break
print("")
print("=============================================")        
print("       ********CATALOGUE********")
print("")
fdr = open("ProductRecord.json",'r')
catalogue1 = fdr.read()
fdr.close()
prod_rec=json.loads(catalogue1)
print("ITEM-ID "+ "NAME " + "PRICE / UNITS "+ "EXPRIREDATE")
for x in prod_rec:
    print(x +"   "+ prod_rec[x]["name"] + "   "+ str(prod_rec[x]["SP"]) +" / "+str(prod_rec[x]["units"])+ "   "+str(prod_rec[x]["exp-date"]))# recording= [json.loads(line) for line in open('record.json', 'r')]
print("")
print("=============================================")      
```

```python
# step2:taking orders and billing and updating the product record
################## FINAL PURCHASING /BILLING AND INVENTORY
import numpy as np
import time
import uuid
ans='y'
cart=[]

q=[]
while ans=='y' or ans=='Y':
    a = str(input("Enter product id:"))   
#     name = str(input("Enter name:"))
    b = int(input("Enter quantity:"))
    topr=time.ctime()         # //time_of_purchase                   
    trans_ID=str(uuid.uuid4())
    if(a in prod_rec.keys()):
        for i in prod_rec.keys():
            if(i==a):
                if(prod_rec[a]["quant"]>=b):
                    prod_rec[a]["quant"]-=b
                    prod_rec[a]["salerec"]["sale"]=True
                    prod_rec[a]["salerec"]["Salequant"]+=b
                    prod_rec[a]["salerec"]["saleTime"].append(topr)
                    prod_rec[a]["salerec"]["transectionID"].append(trans_ID)              
                    cart.append(a)
                    q.append(b)
                   
                    print("ADDED TO YOUR CART")
                    print("")
                    ans = str(input("want to purchase more ?")) 
                    print("")
                else:
                    print("SRY OUT OF STOCK")
                    print("")
                    ans = str(input("want to purchase more ?"))
                    print("")
                    break

    else:
        print("ITEM NOT FOUND")
        print("")
        ans = str(input("want to purchase more ?"))
        print("")
cart      
#update inventory after purchase
js = json.dumps(prod_rec,indent=1)
fd = open("ProductRecord.json",'w')
fd.write(js)
fd.close()


total=0
sum=0
print("")
print("Thank you for Shopping !!")
print("")
print("$$$$$$$$ YOUR BILL $$$$$$$$$$$")
print("")
print("Thank u for purchasing  Mr/Mrs  "+nama)
print("")
print("time of purchase:"+str(topr))
print("transection ID: "+str(trans_ID))
print("")
print("ID  "+" ITEMs "+" "+"QNTY" +" "+" SP"+ " "+ " TOTAL")
l=len(cart)
for i in range(0,l):
    x=cart[i] 
    y=int(q[i])
    sum= y*prod_rec[x]["SP"]
    total=total+sum
    print(x +" "+ prod_rec[x]["name"] + "    "+str(y)+"   "+ str(prod_rec[x]["SP"]) + "   "+str(sum))
print("")
print("========= YOUR BILL =========")
print("TOTAL BILL :  Rs."+ str(total))   
print("$$$$$$$$ YOUR BILL $$$$$$$$$$$")
```

```python
#step3: updating his/her purchase details in his/her customer database
fdcr = open("Cust_Record.json",'r')
r = fdcr.read()
fdcr.close()
cust_rec_up = json.loads(r)


for c in cust_rec_up.keys():
    if(c==currentCustID):
        cust_rec_up[c]["purchase_rec"].append(cart)
        cust_rec_up[c]["topr"].append(topr)
        cust_rec_up[c]["transectio_id"].append(trans_ID)
        cust_rec_up[c]["billAmount"].append(total)
        
# print(cust_rec_up)
xyz = json.dumps(cust_rec_up,indent=1)
f = open("Cust_Record.json",'w')
f.write(xyz)
f.close()
```
